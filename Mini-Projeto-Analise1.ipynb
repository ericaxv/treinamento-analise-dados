## Análise 1

```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
df.head()
```

## Distribuição de Veículos com base no Ano de Registro

```python
# Crie um Plot com a Distribuição de Veículos com base no Ano de Registro
fig, ax = plt.subplots(figsize=(8,6))
sns.distplot(df['yearOfRegistration'], color="#33cc33",kde=True, ax=ax)
plt.xlabel('Ano de Registro', fontsize=15)
plt.ylabel('Densidade(KDE)', fontsize=15)
plt.title('Distribuição de Veículos com base no Ano de Registro', fontsize=15)
plt.show()
```

```python
# Salvando o plot
fig.savefig("plots/Analise1/vehicle-distribution1.png")
```

## Variação da faixa de preço pelo tipo de veículo

```python
df.head()
```

```python
# Crie um Boxplot para avaliar os outliers
fig, ax = plt.subplots(figsize=(10,8))
sns.set_style("whitegrid")
x = df['vehicleType']
y = df['price']
sns.boxplot(x,y)
ax.text(5.25,27000,"Análise de Outliers",fontsize=18,color="r",ha="center", va="center")
plt.xlabel('Tipo de Veículo', fontsize=15)
plt.ylabel('Range de Preço', fontsize=15)
plt.show()
```

```python
# Salvando o plot
fig.savefig("plots/Analise1/price-vehicleType-boxplot1.png")
```

## Contagem total de veículos à venda conforme o tipo de veículo

```python
# Crie um Count Plot que mostre o número de veículos pertencentes a cada categoria 
fig, ax = plt.subplots(figsize=(10,8))
sns.set_style("white")
x = df['vehicleType']
sns.countplot(x, palette="BuPu")
plt.xlabel('Tipo de Veículo', fontsize=15)
plt.ylabel('Total de Veículos para Venda', fontsize=15)
plt.title('Contagem de Veículos à venda conforme o tipo de veículo', fontsize=20)
plt.show()
```

```python
# Salvando o plot
fig.savefig("plots/Analise1/count-vehicleType1.png")
```
